```python
import pandas as pd
from astropy.coordinates import SkyCoord
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy import units as u
```

```python
df = pd.read_csv('NEP_Phase3.txt', delim_whitespace=True)
```

```python
skycoords = SkyCoord(ra = df.RA.values, dec =df.DEC.values, unit = (u.hourangle, u.degree))
```

```python
ra = skycoords.ra.value
dec = skycoords.dec.value
```

```python
pri_1 = dec < 65.7
```

```python
pri_2 = dec > 67
```

```python
pri_21 = (dec >  65.7) & (dec < 66.1)
```

```python
pri_3 = (dec > 66.1) & (dec < 66.2)
```

```python
priority = np.ones(len(ra))
priority[pri_2] = 2
priority[pri_21] = 2
priority[pri_3] = 3
mask = ~(pri_1 | pri_2 | pri_21 | pri_3)
priority[mask] = 4
```

```python
plt.scatter(ra, dec, c = priority)
```

```python
new_df = df.drop(columns = ['PIPRI', 'AZRES'])
```

```python
new_df['PRI'] = priority.astype(int)
```

```python
new_df.to_csv('NEP_2025_Priority.txt', sep = ' ', index = False)
```
